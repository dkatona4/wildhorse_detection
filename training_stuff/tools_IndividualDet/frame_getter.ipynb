```python
import cv2
import csv
import random
import numpy as np 
import os.path


dir_path = '/home/wildhorse_project/Individual_recordings/Villo/'
output_path = '/home/wildhorse_project/detectron_pic/Villo/'


tempo = os.listdir(dir_path)
for i in tempo:
    video_path = dir_path + i
    print(video_path)
    #video_path = '/home/wildhorse_project/Individual_recordings/Emese/Emese_0027.MOV'
    base_video_name = os.path.basename(video_path).split('.')[0]
    vidcap = cv2.VideoCapture(video_path)

    success,image = vidcap.read()

    total_frame = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    random_gen = round(total_frame*0.1)         # always 10 percent
    print(random_gen)
    #random_gen = 200

    rand_array = random.sample(range(0, total_frame), random_gen)
    rand_array.sort()
    count = 0
    row_counter = 0

    while success:
        if rand_array[row_counter] == count:
            filename = output_path + base_video_name + "_" + str(count) + ".jpg"
            cv2.imwrite(filename, image)      
            row_counter += 1
            print(filename," DONE")
        success,image = vidcap.read()
        count += 1
        if row_counter == len(rand_array):
            break
    print("MEGADONE")
```
