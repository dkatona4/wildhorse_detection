```python
import os
import pandas as pd
import sys
import glob

# Must modify
black_path = '/home/wildhorse_project/detectron_pic/black/**/*.csv'
crop_path = '/home/wildhorse_project/detectron_pic/crop/**/*.csv'

def path_check(path, message):
    if glob.glob(path):
        print(message + " path - SUCCESS")
    else:
        print(message + " path - FAILED INVALID PATH: " + path)
        return False
    return True

result = pd.DataFrame()
# Check path definitions
if (path_check(crop_path, "Crop")):
    crop = pd.concat(map(pd.read_csv, glob.glob(crop_path)))
    result = result.append(crop)
if (path_check(black_path, "Black")):
    black = pd.concat(map(pd.read_csv, glob.glob(black_path)))
    result = result.append(black)

# result
    
rows = result.loc[result['Individual'] == 'pinty']
new = rows.groupby(['VideoID','Individual']).size().reset_index(name='counts')
new['Weight'] = new['counts']/len(rows)
new
# max_row = len(rows)
# unique_videoIDs= rows.VideoID.unique()
# for i in unique_videoIDs:
#     new = rows.loc[rows['VideoID'] == i].sample(weight = 0.13)
#     print(str(max_row) + ", " + str(len(new)) + ", " + str(i) ) 

```
