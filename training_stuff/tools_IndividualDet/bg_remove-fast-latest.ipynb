```python
import pickle
import cv2
import os.path
from os import path as p
import matplotlib.pyplot as plt
import numpy as np

def fileopen(filelist, horse_name):
    txt = open(filelist, "r")
    print(horse_name)
    for line in txt:
        rm_bg_and_save(line.strip(),horse_name)

# util method to convert the detectron2 box format
def xyxy_to_xywh(box):
    x1 = int(box[0])
    y1 = int(box[1])
    x2 = int(box[2])
    y2 = int(box[3])
    w = x2-x1
    h = y2-y1
    return [x1,y1,w,h]

def rm_bg_and_save(FILE,horse_name):
    with open(FILE, 'rb') as fi:
        results = pickle.load(fi)
    
    # available classes
    classes = results['classes']
    # Choose the object types you want to extract e.g. "dog"
    class_filter = ["horse"]
    background_color = (0,0,0,255)
    # Cut out masks
    print("Extracting objects...")
    index = 0
    for path in results['instances']:
        tensor_array = results['instances'][path].pred_boxes.tensor.cpu().numpy()
        # Get the rectangle with the largest area
        counter = 0
        max_index = 0
        for i in tensor_array:
            area = (i[2]-i[0])*(i[3]-i[1])
            if counter == 0:
                max = area
                max_index = counter
            elif max < area:
                max = area
                max_index = counter
            counter+=1

        # Path definition
        path_array = path.split('_')
        array_length = len(path_array)
        exact_video_name = path_array[array_length-2]
        exact_frame_index = path_array[array_length-1][:-4]
        
        # Mask definition
        mask = results['instances'][path].pred_masks[max_index]
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)
        
        # Crop around the horse
        box = results['instances'][path].pred_boxes[max_index]
        box = box.tensor.cpu().numpy()[0]
        box = xyxy_to_xywh(box)
        original = img[box[1]:box[1]+box[3], box[0]:box[0]+box[2]]
        crop_img_path = '/home/wildhorse_project/detectron_pic/crop/photo/' + horse_name + '/' + horse_name + '_' + str(exact_video_name) + "_" +str(exact_frame_index) + ".png"
        cv2.imwrite(crop_img_path,original)
        print("CROP - Removed background from '" + path+"'. Saved object in '" + crop_img_path + "")
        
        # Set background color to black
        img[mask.cpu()!=True,:] = background_color
        
        # Cropping image to the size of the objects bounding box
        box = results['instances'][path].pred_boxes[max_index]
        box = box.tensor.cpu().numpy()[0]
        box = xyxy_to_xywh(box)
        img = img[box[1]:box[1]+box[3], box[0]:box[0]+box[2]]
        new_img_path = '/home/wildhorse_project/detectron_pic/black/photo/' + horse_name + '/' + horse_name + '_' + str(exact_video_name) + "_" +str(exact_frame_index) + ".png"       
        cv2.imwrite(new_img_path,img)
        print("BLACK - Removed background from '" + path+"'. Saved object in '" + new_img_path + "")
        
        index+=1

    print("Done...")        
        
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/szazxszep.txt","szazxszep")   # 1
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/thetisz.txt","thetisz")       # 2 
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/panka.txt","panka")           # 3
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/huba.txt","huba")             # 4
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/lilla.txt","lilla")           # 5
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/lantos.txt","lantos")         # 6
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/gerle.txt","gerle")           # 7
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/noci.txt","noci")             # 8
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/vadoc.txt","vadoc")           # 9
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/emese.txt","emese")           # 10
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/hollo.txt","hollo")           # 11
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/honorka.txt","honorka")       # 12
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/luxus.txt","luxus")           # 13
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/marrakesh.txt","marrakesh")   # 14
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/pajzsika.txt","pajzsika")     # 15
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/pinty.txt","pinty")           # 16
# fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/villo.txt","villo")           # 17

fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/szazszorszep_catalog.txt","szazxszep")   # 1
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/thetisz_catalog.txt","thetisz")       # 2 
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/panka_catalog.txt","panka")           # 3
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/huba_catalog.txt","huba")             # 4
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/lilla_catalog.txt","lilla")           # 5
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/lantos_catalog.txt","lantos")         # 6
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/gerle_catalog.txt","gerle")           # 7
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/noci_catalog.txt","noci")             # 8
fileopen("/home/wildhorse_project/detectron_pic/pkl_lists/vadoc_catalog.txt","vadoc")           # 9






```
